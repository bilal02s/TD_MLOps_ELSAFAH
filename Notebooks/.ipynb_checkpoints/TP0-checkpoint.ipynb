```python
x = np.arange(-10, 10.5, 0.5)
y = x**2

plt.plot(x, y)

plt.show()
```

```python
y1 = 2*x + 3 
y2 = -x**2

fig, ax = plt.subplots(1,1)
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_title('Titre')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.grid(True)
```
